# Task1 : Variable and Data Types 

Create a Python script that : 
1. Stores your personal information in variables : 
- Full name (string)
- Age (integer)
- Height in meters (float)
- Favourite programming languages (string)
- Years of Coding Experience (integer)
- Currently learning Python (boolean)
2. Display all this information using f-strings in a nicely formatted way

3. Use `type()` to show the data type of each variable



```python
# taking inputs
name = input("What is your name : ")
age = int(input("What is your age : "))
height = float(input("What is your height : "))
prog_language = input("What is your favourite programming language : ")
year_of_coding_exp = int(input("What is your coding experience in years : "))
learning_python = bool(input("Are you currently learning Python : "))

print(f"""
=== Personal Information ===
    Name : {name}, type : {type(name)}
    Age : {age}, type : {type(age)}
    Height : {height} m, type : {type(height)}
    Programming languages : {prog_language}, type : {type(prog_language)}
    Years of Coding experience : {year_of_coding_exp} yrs, type : {type(year_of_coding_exp)}
    Are you currently learning Python : {learning_python}, type : {type(learning_python)}
=============================
""")
```

# Task 2 : f-strings Practice 

Create a script that : 

1. Stores the following information about a product: 
- Product name : "Mouse"
- Price : 29.99
- Quantity : 5 
- Discount percentage : 10 

2. Calculate : 
- Total before discount (price x quantity)
- Discount amount (total x discount percentage / 100)
- Final total after discount 

3. Display a formatted reciept using f-strings that shows:
- Product Name 
- Unit price 
- Quantity 
- Subtotal 
- Discount percentage and amount 
- Final total 

```python
# taking the inputs 
prod_name = input(f"Name of the Product : ")
price = float(input(f"Price of Product : "))
quantity = int(input(f"Quantity of Product : "))
discount_percent = int(input(f"Discount percent : "))

# claculating 
total_before_discount = price * quantity 
discounted_price = (discount_percent / 100) * (total_before_discount)
total_after_discount = total_before_discount - discounted_price

print(f"""
====== Product Information =======
      Product Name : {prod_name}, type : {type(prod_name)}
      Unit Price : {price}, type : {type(price)}
      Quantity : {quantity}, type : {type(quantity)}
      Subtotal : {total_before_discount}, type : {type(total_before_discount)}
      Discount percentage : {discount_percent}, type : {type(discount_percent)}
      Discount price : {discounted_price}, type : {type(discounted_price)}
      Final Total : {total_after_discount}, type : {type(total_after_discount)}
==================================
""")
```

# Task 3: User Input and Type Conver Task 3: 

Write a program that : 
- Asks user their age 
- Asks if they have a driver's license 
- Uses comparison and logical operators to determine

    - Can they drive? (age >= 18 AND has license)
    - Are they a senior? (age >= 65)
    - Are they a teenager? (age >= 13 AND age < 20)
    - Can they vote? (age >= 18)
- Display the results in a clear format

```python
age = int(input(f"What is your age : "))
has_license = input(f"Do you have driving license : ").lower() == "true"

if age >= 18 and has_license:
    print(f"You can drive!")

if age >= 65:
    print(f"You are a senior person")

if age >= 13 and age < 20:
    print(f"You are a teenager.")
    
if age >= 18:
    print(f"You can vote!!!!")
```
