##### **f-strings**
Allows you to insert variables as placeholder withing a string

```python
# Before f-strings we need to add variables like this 
name = "Alice"
age = 30 

message = "Hey, my name is " + name + ", my age is " + str(age)
print(message)
```

```python
# after using f-strings
print(f"Hey, My name is {name}, my age is {age}")
```

##### **User Input**  
The `input` function pauses your program and waits for user to type something and press Enter.

```python
name = input('Write your name : ')
print(f"Hello {name} !")
```

IMPORTANT : `input` always return a string, even if the user types a number!

```python
age = input('How old are you? ')
print(type(age))
```

```python
# we can convert the input to a number if required 
age = input('How old are you? ')
age = int(age)
print(type(age))
```

##### **Operators: Tools for working with Data**

Arithmetic Operators

```python
a = 10
b = 3 

print(a + b)
print(a - b)
print(a * b)
print(a/b)
print(f"Only gives integer part after division : {a//b}")
print(f"Gives the remainder : {a%b}") 
print(f"a to the power of b : {a ** b}")
```

Comparision Operators 
- Compares value and return `True` and `False`

```python
a = 10
b = 5

print(a == b) # checks for equality 
print(a != b) # Not equals to 
print(a > b) # a greater than b
print(a < b) # a less than b 
print(a >= b) # a greater than and equals to 
print(a <= b) # less than and equals to
```

#### **Logical operators**

AND : (both needs to be `True`)
- `True` and `True` = `True`
- `True` and `False` = `False` 
- `False` and `False` = `False` 
 
OR : (atleast one needs to be `True`)
- `True` or `True` = `True`
- `True` or `False` = `True` 
- `False` or `False` = `False` 

NOT : (negates the value)
- not `True` = `False`
- not `False` = `True`

```python
age = 25 
has_license = True

# AND operator 
if age >= 18 and has_license:
    print(f"You can drive!")

# OR operator 
if age < 18 or age > 65:
    print(f"You need a driver then!")

# NOT 
if not has_license:
    print(f"You need a license!")
```
