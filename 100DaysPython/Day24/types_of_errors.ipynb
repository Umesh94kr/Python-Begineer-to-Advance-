#### **Try and Except block**

try : Something that might cause exception 

except : Do this if there was an exception 

else : Do this if there was no exception 

finally : Do this no matter what happens


```python
try:
    with open('file1.txt', 'r') as f:
        content = f.read()
        print(content)
except Exception as e:
    print(f"File not exists : {e}")
```

```python
try:
    list1 = ['apple', 'banana', 'grapes']
    f1 = list1[3]
except IndexError as e:
    print(f"Key not found. Error : {e}")
```

#### **Raising own errors**

```python
height = float(input("Height : "))
weight = float(input("Weight : "))

bmi = round(weight / (height * height), 2)

if height > 3:
    raise ValueError("Human height can't be more than 3 meters.")

print(f"BMI : {bmi}")
```
